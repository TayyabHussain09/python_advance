```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Luke_hair_loss_documentation.csv")
```

```python
df
```

```python
#selecting of the input and output variable
X=df[["coffee_consumed","brain_working_duration","stay_up_late"]]
y=df["hair_loss"]
```

```python
# machine learniong algorithm
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier().fit(X,y)
```

```python
model.predict([[3,3,3]])
```

```python
# how to meaure the accuracy of our model
#split data into the test and train(80/20)
from  sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train , y_test = train_test_split(X,y, test_size = 0.2) # 80% trainign remaing testing

# create a model
model.fit(X_train,y_train)

predicted_values = model.predict(X_test)

predicted_values = model.predict(X_test)
predicted_values

print("-------------------------------------------")

# checking the score of the predicted values 


actual_values = y_test

score = accuracy_score(actual_values, predicted_values)
print("score is", score) # ohh wow





```

```python
# graph
from sklearn import tree
model = DecisionTreeClassifier().fit(X,y)

tree.export_graphviz(model, out_file = "hairloss.dot",feature_names=["cofee_consumed","brain_working_duration",'stay_up_late'],class_names=sorted(y.unique()),label="all",rounded=True,filled=True)
```

```python
# how to train and save yr model
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
import joblib

model = DecisionTreeClassifier().fit(X,y)

joblib.dump(model, "hair_loss.joblib")
 
```

```python
# how to run the store model

model = joblib.load("hair_loss.joblib")
model
```
