# Hands on Example

```python
# important dataset

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


# importting the dataset

dataset = pd.read_csv("http://s3.us-west-2.amazonaws.com/public.gamelab.fun/dataset/position_salaries.csv")
```

```python
dataset.all
```

```python
X = dataset.iloc[:, 1:2].values
y= dataset.iloc[:, 2].values
```

```python
# splitting the dataset into the training and testing

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
# fitting linear regression to the dataset

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
# visualizing the linear regression results
def viz_linear():
    plt.scatter(X,y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
viz_linear()
```

```python
# fitting polynomial regressio0n to the dataset

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
pol_reg = LinearRegression()
pol_reg.fit(X_poly,y)

# visualizing the polyomial regression results

def viz_polynomial():
    plt.scatter(X,y, color='red')
    plt.plot(X, pol_reg.predict(poly_reg.fit_transform(X)), color='blue')
    plt.title('Truth or Bluff(Linear Regression)')
    plt.xlabel('position level')
    plt.ylabel('salary')
    plt.show()
    return
viz_polynomial()
```

```python
# predicting the new result with linear regression
pred_linear = lin_reg.predict([[11]])
pred_linear
```

```python
pred_polynomial = pol_reg.predict(poly_reg.fit_transform([[11]]))
pred_polynomial
```

```python
print('Linear Regression Results   = ', pred_linear)
print('----------------------')
print('Polynomial Regression Results   = ', pred_polynomial)
```
