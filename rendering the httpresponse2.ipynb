```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.imf.org/en/home"
```

```python
data = requests.get(url)
```

```python
model = data.content
```

```python
apply = BeautifulSoup(model, 'html.parser')
```

```python
data.close()
```

```python
c=apply.contents
c
```

```python
print(apply.prettify())
```

```python
for href in apply.findAll('a', href=True):
    print(href["href"])
```
