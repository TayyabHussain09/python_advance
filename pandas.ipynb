```python
from turtle import *

bgcolor('black')
speed(0)
hideturtle()
for i in range(120):
    color('red')
    circle(i)
    color('orange')
    circle(i*0.8)
    right(3)
    forward(3)
done()
```

```python
from turtle import *

bgcolor('black')
speed(1500088686000)
hideturtle()
for i in range(100072835427834500000):
    color('yellow')
    circle(i)
    color('green')
    circle(i*3)
    right(2)
    forward(2)
    left(i*3*2)
    color('red')
    circle(i*876875756)
    color('blue')
    circle(i*3657657)
    color('green')
    circle(i*300000)
    color('cyan')
    color('black')
    color('white')
    color('navy')
    circle(i)
    round(i*10)
done()
```

```python
import turtle as t
import colorsys
t.width(3)
t.tracer(10)
t.bgcolor('black')
h=0
n=50
for i in range(200):
  c=colorsys.hsv_to_rgb(h,1,0.8)
  h+=1/n
  t.pencolor(c)
  t.circle(i,90)
  t.forward(i)
  t.right(270)
  t.circle(i,270)
  t.forward(i)
  t.right(180)
t.done
```
